```python
import random
import matplotlib.pyplot as plt

def coinflip(numberofflips):
    coin = ["Heads", "Tails"]
    flips = [random.choice(coin) for item in range(numberofflips)]
    return flips

def ratio(flips):
    head = flips.count("Heads")
    tail = flips.count("Tails")
    return (head, tail)
def flipping(grow):
    heads_number = []
    tails_number = []
    for item in grow:
        heads_number.append(ratio(coinflip(item))[0])
        tails_number.append(ratio(coinflip(item))[1])
    return (sorted(tails_number), sorted(heads_number))
def means():
    number_of_heads = sum(heads_number)
    number_of_tails = sum(tails_number)
    total = number_of_tails + number_of_heads
    print(total)
    rati = number_of_heads/total
    return rati
```

```python
grow = [2**0, 2**1, 2**2, 2**4, 2**8, 2**16]
tup = flipping(grow)
x1 = grow
y1 = tup[1]
z1 = tup[0]
plt.plot(y1)
plt.plot(z1)
plt.xticks(range(6), x1)
plt.xlabel("Growing By")
plt.ylabel("Frequency")
plt.title("Coin")
plt.legend(["Tails Number", "Heads Number"], loc='upper left')
plt.show()
```
